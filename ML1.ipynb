```python
import numpy as np
#define array
arr=np.array([5,6,11])
#calculate mean
mean=np.mean(arr)
print("mean=",mean)
```

```python
from scipy import stats
arr=[1,2,3,4,4]
mode_result=stats.mode(arr)#calculation
print("mode=",mode_result.mode)#removes[0]
```

```python
import numpy as np
arr=np.array([1,2,3,4,5,6])#define array
median=np.median(arr)#calculation
print("median=",median)#result
```

```python
import statistics
arr=[1,2,3,4,5]
var=statistics.variance(arr)
print("variance=",var)
```

```python
import statistics
arr=[1,2,3,4,5]
std=statistics.stdev(arr)
print("standard deviation=",std)
```

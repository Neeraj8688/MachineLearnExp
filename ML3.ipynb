```python
import numpy as np
import pandas as pd
df=pd.read_csv('/content/tested.csv')

```

```python
df.info()

df=df.drop(columns=['PassengerId','Name','Ticket','Cabin'])
print("Columns after dropping irrelevant ones:",df.columns)






df['Age'].fillna(df['Age'].median(), inplace=True)
df['Fare'].fillna(df['Fare'].median(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
print("Remaining missing values:\n", df.isnull().sum())







age_bins = [0, 12, 18, 50, 100]
age_labels = ['Child', 'Teenager', 'Adult', 'Senior']
df['AgeGroup'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels)
fare_bins = [-1, 50, 100, 250, 600]
fare_labels = ['Low', 'Medium', 'High', 'Very High']
df['FareRange'] = pd.cut(df['Fare'], bins=fare_bins, labels=fare_labels)
print(df[['Age', 'AgeGroup', 'Fare', 'FareRange']].head())







Q1 = df['Fare'].quantile(0.25)
Q3 = df['Fare'].quantile(0.75)
IQR = Q3 - Q1
df = df[(df['Fare'] >= Q1 - 1.5 * IQR) & (df['Fare'] <= Q3 + 1.5 * IQR)]
print("DataFrame shape after outlier removal:", df.shape)


```
